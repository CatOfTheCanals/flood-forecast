```python
import numpy as np
import pandas as pd
```

```python
# Import the excel file and call it xls_file
xls_file = pd.ExcelFile('daily_met.xls')
xls_file
```

```python
# View the excel file's sheet names
xls_file.sheet_names
df = xls_file.parse('evap')
df.head()
```

```python
df['date_evap'] = pd.to_datetime(df['date_evap'],format='%Y-%m-%d')

for daysDelta in range(-20,-6): #CHANGE HERE
        key = 't_evap'+str(daysDelta)
        df[key] = np.nan
df.head()
```

```python
for index, row in df.iterrows():
    for daysDelta in range(-20,-6): #loops through days CHANGE HERE
        key = 't_evap'+str(daysDelta)

        # will be true if the difference of days is the one you are looking for
        booleanTimeDelta = ((df.loc[:,'date_evap'] -row['date_evap']).dt.days == daysDelta) 

        # if any are true find them and assign it  
        if any(booleanTimeDelta):
            df.loc[index:index+1,key] = df.loc[booleanTimeDelta,'evap'].values
```

```python
print(df.head(n=20))
```

```python
# Import the excel file and call it xls_file
xls_file_2 = pd.ExcelFile('dates_aquisition.xls')
xls_file_2
xls_file_2.sheet_names
```

```python
df_dates = xls_file_2.parse('Dates_aquisition')
df_dates.head()
```

```python
merge_df = pd.merge(df, df_dates, left_on = ['date_evap'], right_on = ['date_aqui'])
```

```python
merge_df.drop('date_aqui', axis=1, inplace=True)
merge_df.drop(0, axis=0, inplace=True)
merge_df.head()
```

```python
merge_df.to_csv('evap_dataframe.csv')
```
